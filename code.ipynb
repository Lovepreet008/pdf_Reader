```python

def pdf_to_dataframe(pdf_path):
    # Use tabula to extract tables from the PDF
    tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True, stream=True)
    # tables = camelot.read_pdf(pdf_path)
    
    # Combine the extracted tables into a single DataFrame
    if len(tables) > 0:
        df = pd.concat(tables, ignore_index=True)
        return df
    else:
        print("No tables found in the PDF.")
        return None

# Replace 'your_pdf_file.pdf' with the actual path to your PDF file
pdf_path='Resources/M2TST2E12M1K02967.pdf'

# Call the function to convert PDF data into a DataFrame
result_df = pdf_to_dataframe(pdf_path)

# Display the resulting DataFrame
if result_df is not None:
    print(result_df)
```
